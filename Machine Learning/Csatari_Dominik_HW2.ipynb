```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np




data=pd.read_csv('heart_disease.csv')
data.hist()
labels=data.values[:,-1]
labels[labels>1]=1
labels=labels.astype(int)

data=data.values[:,:-1]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.33, random_state=42)

#basic linear regression with test and train sets.
clf = LogisticRegression(random_state=0).fit(x_train, y_train)
clf.predict(x_test)

clf.predict_proba(x_test)

clf.score(x_test, y_test)


```

**BUILT-IN POLY**

```python
from sklearn.preprocessing import PolynomialFeatures
# define the dataset

# perform a polynomial features transform of the dataset

trans = PolynomialFeatures(degree=2)
data_trans = trans.fit_transform(data)

x_train_tr, x_test_tr, y_train_tr, y_test_tr = train_test_split(data_trans, labels, test_size=0.33, random_state=42)

#basic linear regression with test and train sets.
clf = LogisticRegression(random_state=0).fit(x_train_tr, y_train_tr)
clf.predict(x_test_tr)

clf.predict_proba(x_test_tr)

clf.score(x_test_tr, y_test_tr)


```

**Own function**

```python

def polynomial_features(X, degree):
    
    n_samples, n_features = X.shape
    
    # a0= konstans
    X_poly = np.ones((n_samples, n_features * degree + 1))

    for i in range(n_features):
        for d in range(1, degree + 1):
            X_poly[:, i*degree + d] = np.power(X[:, i], d)

    return X_poly
```

```python

data_trans_2 = polynomial_features(data,2)

x_train_trr, x_test_trr, y_train_trr, y_test_trr = train_test_split(data_trans_2, labels, test_size=0.33, random_state=42)

clf = LogisticRegression(random_state=0).fit(x_train_trr, y_train_trr)
clf.predict(x_test_trr)

clf.predict_proba(x_test_trr)

clf.score(x_test_trr, y_test_trr)
```
